# 1. Importing matplotlib and some necessary libraries

```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt

%matplotlib inline
# so that the graph generates below the coding cell not in a different window
```

#### 1.1 Generating Sin and Cos Graph using Matplotlib

```python
x = np.linspace(0, 6*np.pi, 100)

#Defining the wave
sin_y = np.sin(x)
cos_y = np.cos(x)

# Plotting the wave
plt.plot(x, sin_y, label = 'Sine')
plt.plot(x, cos_y, label = 'Cosine')

plt.legend(loc = 'upper right')  # to display the label

plt.title("Sin and Cos Wave Graph")

plt.xlabel("x")
plt.ylabel("sin(x) or cos(x)")

plt.show()
```

```python
x = np.linspace(0, 3*np.pi, 100)

#Defining the wave
sin_y = np.sin(x)
cos_y = np.cos(x)

# Plotting the wave
plt.plot(x, sin_y,  '*')
plt.plot(x, cos_y,  '--')

plt.title("Sin and Cos Wave Graph")

plt.xlabel("x")
plt.ylabel("sin(x) or cos(x)")

plt.show()
```

# 2. Setting Styles to th Graph

#### 2.1 One way to use styles is use `plt.style.use('style_name')`. List of available styles is:

```python
print(plt.style.available)
```

```python
plt.style.use('bmh')

plt.plot(x, sin_y, label = 'Sine')
plt.plot(x, cos_y, label = 'Cosine')

plt.title("Sin and Cos Wave Graph")

plt.legend()
plt.xlabel("x")
plt.ylabel("sin(x) or cos(x)")

plt.show()
```

#### 2.2 We can also increase size of the plot

`plt.figure(figsize = (length,height))`

```python
plt.figure(figsize = (10,5))

plt.style.use('seaborn-v0_8-paper')

plt.plot(x, sin_y, label = 'Sine')
plt.plot(x, cos_y, label = 'Cosine')

plt.title("Sin and Cos Wave Graph")

plt.legend()
plt.xlabel("x")
plt.ylabel("sin(x) or cos(x)")

plt.show()
```

# 3. Saving the Plots in our Local Machine

```python
plt.figure(figsize = (18,5))

plt.style.use('seaborn-v0_8-paper')


fig = plt.figure()

plt.plot(x, sin_y, label = 'Sine')
plt.plot(x, cos_y, label = 'Cosine')

plt.title("Sin and Cos Wave Graph")

plt.legend()
plt.xlabel("x")
plt.ylabel("sin(x) or cos(x)")

plt.show()

fig.savefig('Sine and Cosine.png')
```
