# 1. Importing matplotlib and some necessary libraries

```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
import sklearn
%matplotlib inline
# so that the graph generates below the coding cell not in a different window
```

#### 1.1 Generating Sin and Cos Graph using Matplotlib

```python
x = np.linspace(0, 6*np.pi, 100)

#Defining the wave
sin_y = np.sin(x)
cos_y = np.cos(x)

# Plotting the wave
plt.plot(x, sin_y, label = 'Sine')
plt.plot(x, cos_y, label = 'Cosine')

plt.legend(loc = 'upper right')  # to display the label

plt.title("Sin and Cos Wave Graph")

plt.xlabel("x")
plt.ylabel("sin(x) or cos(x)")

plt.show()
```

```python
x = np.linspace(0, 3*np.pi, 100)

#Defining the wave
sin_y = np.sin(x)
cos_y = np.cos(x)

# Plotting the wave
plt.plot(x, sin_y,  '*')
plt.plot(x, cos_y,  '--')

plt.title("Sin and Cos Wave Graph")

plt.xlabel("x")
plt.ylabel("sin(x) or cos(x)")

plt.show()
```

# 4. Different Types of Plots

#### 4.1 Line Plot

* A line plot is a chart that displays data points connected by straight lines.

* It is great for `**showing trends**` or `**changes over time**`.

* In Data Science, it is most commonly used for `time-series data` (e.g., stock prices, sales over months, temperature changes).

* It clearly shows upward or downward `trend`, `patterns`, and `fluctuations`.

* Use a line plot when your data is continuous and you want to highlight the `relationship` or `progression` between values.

* It helps in spotting trends, seasonality, and anomalies at a glance.

```python
x = np.linspace(0, 4*np.pi,100  )
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize = (15,5))
plt.style.use('Solarize_Light2')

plt.plot(x, y_sin, color = 'red', linewidth = 2, label  = 'Sine')
plt.plot(x, y_cos, color = 'green', linewidth = 2, label = 'Cos')

plt.legend(loc = 'upper right')

plt.xlabel("x")
plt.ylabel("sin(x) or cos(x)")

plt.title("Sin and Cos Wave Graph")

plt.show()
```

##### 4.1.3 Adjusting the Plot Axes

1. Can be done using `plt.xlim` for x-axis.
2. `plt.ylim` for y-axis


```python
x = np.linspace(0, 4*np.pi, 100)

plt.xlim(0,15)
plt.ylim(-2,2)

sin_y = np.sin(x)

plt.plot(x, sin_y, label = 'Sine')


plt.title("Sin Wave Graph")

plt.xlabel("x")
plt.ylabel("sin(x)")

plt.show()
```

can also be done using `plt.axis([x.lim,y.lim])`

```python
x = np.linspace(0, 4*np.pi, 100)

plt.axis([0,15,-2,2])
# plt.xlim(0,15)
# plt.ylim(-2,2)

sin_y = np.sin(x)

plt.plot(x, sin_y, label = 'Sine')


plt.title("Sin Wave Graph")

plt.xlabel("x")
plt.ylabel("sin(x)")

plt.show()
```
